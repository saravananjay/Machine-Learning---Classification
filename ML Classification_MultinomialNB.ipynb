```python
import pandas as pd
dataset = pd.read_csv('heart_disease.csv')
dataset
```

```python
dataset = pd.get_dummies(dataset, drop_first=True)
dataset
```

```python
dataset.Target.value_counts()
```

```python
dataset.columns
```

```python
X = dataset[['Age', 'BP', 'Cholestrol', 'Pulse', 'Sex_male',
       'restecg_Normal', 'Excersise_yes']]
```

```python
y= dataset['Target']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
from sklearn.naive_bayes import MultinomialNB

classifier = MultinomialNB()

classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)

from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)

from sklearn.metrics import classification_report
clf_report = classification_report(y_test, y_pred)
print(clf_report)
```

```python
def get_int_input(prompt):
	while True:
		value = input(prompt)
		if value.strip() == "":
			print("Input cannot be empty. Please enter a value.")
			continue
		try:
			return int(value)
		except ValueError:
			print("Invalid input. Please enter an integer.")

age_input = get_int_input("Enter Age: ")
bp_input = get_int_input("Enter BP: ")
chol_input = get_int_input("Enter Cholestrol: ")
pulse_input = get_int_input("Enter Pulse: ")
sex_male_input = get_int_input("Sex Male 0 or 1: ")
restecg_normal_input = get_int_input("Rest ECG Normal 0 or 1: ")
excersise_yes_input = get_int_input("Excersise Yes 0 or 1: ")
```

```python
classifier_predict = classifier.predict([[age_input, bp_input, chol_input, pulse_input, sex_male_input, restecg_normal_input,excersise_yes_input]])
if classifier_predict[0] == 1:
    print("You have heart disease")
else:
    print("You do not have heart disease")
```
